## 3. Training on Atari with Images

```
# Create environment
env = make_atari_env('FreewayNoFrameskip-v0', seed=42, num_env=4)
env = VecFrameStack(env, n_stack=4)
```

```
env.observation_space
```

```
experiment = 'ppo2_freeway_baseline_run_3'

#Training model
# model = PPO2(CnnPolicy, env, noptepochs=3, cliprange=0.1,vf_coef=1,verbose=1, tensorboard_log=logsPath+"ppo2_tensorboard_experiments/")
model = PPO2(CnnPolicy, env, verbose=1, tensorboard_log=logsPath+"ppo2_tensorboard_experiments/")
print('learning...')
model.learn(total_timesteps=1000000)
model.save(modelsPath+''+experiment)
model.save(modelsPath+''+experiment, cloudpickle=True)
```

## 4. Training on Atari with RAM

```
# Create environment
env = make_atari('Freeway-ramNoFrameskip-v0')
# env = make_atari_env('Freeway-ramNoFrameskip-v0', seed=42, num_env=4)
```

```
env.observation_space
```

```
experiment = 'ppo2_freeway_ram_baseline_mlp_gamma_90_run_2'

#Training model
model = PPO2(MlpPolicy, env, gamma=0.90, verbose=1,tensorboard_log=logsPath+"ppo2_tensorboard_ram_experiments/")
print('learning...')
model.learn(total_timesteps=400000)
model.save(modelsPath+''+experiment)
model.save(modelsPath+''+experiment, cloudpickle=True)
```
